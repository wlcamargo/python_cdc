## Start Incremental Update

```python
source = pd.read_sql_query(""" 
SELECT 
      [ID]
      ,[Coluna1]
      ,[Coluna2]
      ,[Coluna3]
      ,[Coluna4]
      ,[Coluna5]
      ,[Coluna6]
      ,[Coluna7]
      ,[Coluna8]
      ,[Coluna9]
      ,[Coluna10]
      ,[Coluna11]
      ,[Coluna12]
      ,[Coluna13]
      ,[Coluna14]
      ,[Coluna15]
      ,[Coluna16]
      ,[Coluna17]
      ,[Coluna18]
      ,[Coluna19]
      ,[Coluna20]
     
  FROM [TabelaExemplo]
  WHERE ID >= 900000
                           
""", src_conn)
source
```

```python
target = pd.read_sql('Select * from public."stg_IncrementalLoadTest"', engine)
target
```

```python
target.apply(tuple,1)
```

```python
source.apply(tuple,1).isin(target.apply(tuple,1))
```

```python
# Get changes
changes = source[~source.apply(tuple,1).isin(target.apply(tuple,1))]
changes
```

```python
# Get modified rows
modified = changes[changes.ID.isin(target.ID)]
modified
```

```python
# Get new records
inserts = changes[~changes.ID.isin(target.ID)]
inserts
```

```python
def update_to_sql(df, table_name, key_name):
    a = []
    table = table_name
    primary_key = key_name
    temp_table = f"{table_name}_temporary_table"
    
    for col in df.columns:
        if col == primary_key:
            continue
        a.append(f'"{col}"=s."{col}"')
    
    # Create temporary table
    df.to_sql(temp_table, engine, if_exists='replace', index=False)
    
    # Construct and execute the update statement
    update_stmt_1 = f'UPDATE public."{table}" f '
    update_stmt_2 = "SET "
    update_stmt_3 = ", ".join(a)
    update_stmt_4 = f' FROM public."{table}" t '
    update_stmt_5 = f' INNER JOIN (SELECT * FROM public."{temp_table}") AS s ON s."{primary_key}"=t."{primary_key}" '
    update_stmt_6 = f' WHERE f."{primary_key}"=s."{primary_key}" '
    update_stmt_7 = update_stmt_1 + update_stmt_2 + update_stmt_3 + update_stmt_4 + update_stmt_5 +  update_stmt_6 +";"
    
    with engine.begin() as cnx:
        cnx.execute(text(update_stmt_7))
```

```python
# call function update data		
update_to_sql(modified, "stg_IncrementalLoadTest", "ID")
```

```python
# insert new rows into the destination table
inserts.to_sql(tbl_name, engine, if_exists='append', index=False)
```
